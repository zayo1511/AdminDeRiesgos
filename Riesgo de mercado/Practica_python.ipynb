```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
```

```python
St = yf.download(['PLTR', 'MELI'], start='2020-01-01')['Close'].dropna()
St
```

```python
plt.figure(figsize=(12,6))
plt.hist(St, label=St.columns, bins=50)
plt.legend()
plt.show()
```

```python
rt = St.pct_change().dropna()

plt.figure(figsize=(12,6))
plt.hist(rt, label=rt.columns, bins=50)
plt.legend()
plt.show()
```

```python
rt.mean()*252*100
```

```python
rt.std()*252**(1/2)*100
```

```python
rt.skew()
```

```python
rt.kurt()
```
