## <font color = "white"> Parte 1. Selección de Activos <font color = "white">

+ Seleccionar de 6 a 8 activos financieros que coticen en bolsa, sientanse libres de utilizar ETF's, bonos, derivados o acciones.  


+ Justificar la selección de cada uno de los activos con métricas financieras, con Análisis Fundamental (PER, BPA, EBITDA, Beta). En esta parte se tiene que justificar la selección de la mejor manera posible, incluyendo todo lo que debamos saber acerca del instrumento, es decir, hacer una investigación acerca de las características de las empresas, expectativas, planes a futuro.

    
Ejemplo: Lo ideal sería seleccionar activos con, por ejemplo PER bajo, pero, si se quiere seleccionar NVIDIA, comentar todo lo que tiene planeado NVIDIA para el futuro, además de la perspectiva de mercado para la empresa, lo que causa una sobrevaluación de mercado.
    

+ Comentar acerca de la coyuntura económica (checar papers de JP Morgan, Blackrock compartidos).  
    

+ Se debe comentar el horizonte de inversión recomendado y el perfil de inversión al que se dirige nuestro portafolio.  
    
    
+ En caso de que se utilicen bonos, comentar sobre el riesgo de crédito del emisor de manera cualitativa, la calificación crediticia y leer el indentiture del bono.   
    

**Productos Esperados**: Se muestran los activos elegidos, cada uno con sus respectivas justificaciones y características.
